```python
class BestOptimizer1(Optimizer):
    def __init__(self, params, lr=0.001, beta1=0.95, beta2=0.999, epsilon=1e-8,
                 weight_decay=0.005, grad_clip_val=1.0, x_clip_val=10.0):
        defaults = dict(lr=lr, beta1=beta1, beta2=beta2, epsilon=epsilon,
                        weight_decay=weight_decay, grad_clip_val=grad_clip_val,
                        x_clip_val=x_clip_val)
        super().__init__(params, defaults)

    @torch.no_grad()
    def step(self, closure=None):
        loss = None
        if closure is not None:
            loss = closure()

        for group in self.param_groups:
            lr = group['lr']
            beta1 = group['beta1']
            beta2 = group['beta2']
            epsilon = group['epsilon']
            weight_decay = group['weight_decay']
            grad_clip_val = group['grad_clip_val']
            x_clip_val = group['x_clip_val']

            for p in group['params']:
                if p.grad is None:
                    continue

                state = self.state[p]
                if 'v' not in state:
                    state['v'] = torch.zeros_like(p)

                v = state['v']
                g = p.grad

                neg_grad_clip = -grad_clip_val
                g_clipped = torch.clamp(g, neg_grad_clip, grad_clip_val)
                one_minus_beta1 = 1 - beta1
                v_decayed_term = beta1 * v
                g_momentum_contrib = one_minus_beta1 * g_clipped
                v_new = v_decayed_term + g_momentum_contrib
                v_new_abs = v_new.abs()
                softplus_v_new_abs = torch.nn.functional.softplus(v_new_abs)
                g_clipped_abs = g_clipped.abs()
                exp_g_clipped_abs = torch.exp(g_clipped_abs)
                max_adaptive_term = torch.maximum(softplus_v_new_abs, exp_g_clipped_abs)
                adaptive_denom = max_adaptive_term + epsilon
                adaptive_lr_factor = lr / adaptive_denom
                momentum_step = v_new * adaptive_lr_factor
                x_sign_for_wd = torch.sign(p)
                wd_x_term = weight_decay * x_sign_for_wd
                total_update_direction = momentum_step + wd_x_term
                x_candidate = p - total_update_direction
                neg_x_clip_val_final = -x_clip_val
                x_clipped_first = torch.clamp(x_candidate, neg_x_clip_val_final, x_clip_val)
                x_new = torch.tanh(x_clipped_first)

                p.copy_(x_new)
                state['v'] = v_new

        return loss
```

## Experiments

### Sphere

```python
sphere5 = Sphere(dim=5)
sphere25 = Sphere(dim=25)
sphere100 = Sphere(dim=100)

optim_schemas = [
    OptimizerSchema(optimizer_class=optim.SGD, params={"lr": 0.1}),
    OptimizerSchema(optimizer_class=optim.Adam, params={"lr": 0.1}),
    OptimizerSchema(optimizer_class=optim.RMSprop, params={"lr": 0.1}),
    OptimizerSchema(optimizer_class=BestOptimizer1, params={"lr": 0.1}),
]

comp1 = Comparision(
    sphere5, optim_schemas, epochs=50, repetitions=10, seed=42, log_freq=10,
    results_dir="experiment_result/gemini_first"
)

comp2 = Comparision(
    sphere25, optim_schemas, epochs=50, repetitions=10, seed=42, log_freq=10,
    results_dir="experiment_result/gemini_first"
)

comp3 = Comparision(
    sphere100, optim_schemas, epochs=50, repetitions=10, seed=42, log_freq=10,
    results_dir="experiment_result/gemini_first"
)

comp1.run()
comp2.run()
comp3.run()
```

```python
def get_files_from_dir(directory: str, ext: str = "*") -> List[Path]:
    directory_path = Path(directory)
    if not directory_path.exists() or not directory_path.is_dir():
        raise ValueError(f"Katalog {directory} nie istnieje lub nie jest katalogiem.")

    files = list(directory_path.glob(f"*.{ext}" if ext != "*" else "*"))
    return files
```
